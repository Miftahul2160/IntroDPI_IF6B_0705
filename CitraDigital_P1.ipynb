```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Baca gambar dari file
img_bgr = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Citra Digital/P1/kontruksi.jpg')

# Cek apakah gambar berhasil dibaca
if img_bgr is None:
    print("Error: Gambar tidak ditemukan atau gagal dibaca!")
else:
    # Resize agar tidak terlalu besar
    img_bgr = cv2.resize(img_bgr, (300, 200))

    # Cek tipe data
    print("Tipe data:", type(img_bgr))
    print("Shape    :", img_bgr.shape)
    print("Dtype    :", img_bgr.dtype)

    print()
    print("Artinya:")
    print(f"  Tinggi  = {img_bgr.shape[0]} piksel")
    print(f"  Lebar   = {img_bgr.shape[1]} piksel")
    print(f"  Channel = {img_bgr.shape[2]} (BGR)")

    # Konversi BGR ke RGB untuk matplotlib
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

    print("Total piksel:", img_bgr.shape[0] * img_bgr.shape[1])

    # Tampilkan gambar
    plt.figure(figsize=(4, 6))
    plt.imshow(img_rgb)
    plt.title("foto Gambar Kontruksi \n (sudah dikonversi BGR -> RGB)")
    plt.axis('off')
    plt.show()
```

```python
# Konversi ke grayscale
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

# Ambil area kecil 8x8 piksel dari tengah gambar
area_kecil = img_gray[192:200, 100:108]

print("=== MATRIKS f(x,y) — Area 8x8 Piksel ===")
print("(Setiap angka = nilai intensitas 1 piksel (0=hitam, 255=putih))")
print()
print(area_kecil)
print()
print("Contoh baca:", "f(180, 200) =", img_gray[199, 100])
print("Contoh baca:", "f(181, 201) =", img_gray[198, 101])

# ================== Tampilkan area kecil ==================
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Kiri: gambar grayscale penuh
axes[0].imshow(img_gray, cmap='gray', vmin=0, vmax=255)
axes[0].set_title("Gambar Grayscale\n(setiap piksel = 1 angka)")
axes[0].axis('off')

# Tanda area yang kita zoom
import matplotlib.patches as patches
rect = patches.Rectangle((150, 100), 8, 8, linewidth=2, edgecolor='red', facecolor='none')
axes[0].add_patch(rect)
axes[0].text(150, 95, "area ini di-zoom", color='red', fontsize=8)

# Kanan: tampilan area kecil + angka
axes[1].imshow(area_kecil, cmap='gray', vmin=0, vmax=255)

for i in range(8):
    for j in range(8):
        nilai = area_kecil[i, j]
        warna_teks = 'white' if nilai < 128 else 'black'
        axes[1].text(j, i, str(nilai), ha='center', va='center',
                     fontsize=9, color=warna_teks, fontweight='bold')

axes[1].set_title("Zoom 8x8 + nilai tiap piksel")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
bgr_area_kecil = img_bgr[100:108, 175:183]

print("Nilai MATRIKS")
print()
print(bgr_area_kecil)
print()


rgb_area_kecil = cv2.cvtColor(bgr_area_kecil, cv2.COLOR_BGR2RGB)

total_pixel = img_bgr.shape[0] * img_bgr.shape[1]
ukuran_bytes = img_bgr.nbytes
ukuran_mb = ukuran_bytes / (1024 ** 2)

fig_2, axes_2 = plt.subplots(1, 2, figsize=(16, 8))

axes_2[0].imshow(img_rgb)
axes_2[0].set_title('Gambar RGB\n(setiap piksel = 3 angka)')
axes_2[0].axis('off')

rect = patches.Rectangle((175, 100), 8, 8, linewidth=2, edgecolor='blue', facecolor='none')
axes_2[0].add_patch(rect)
axes_2[0].text(175, 95, 'area zoom', color='blue', fontsize=10, fontweight='bold')


axes_2[1].imshow(rgb_area_kecil)
for i in range(8):
  for j in range(8):
    nilai_bgr = bgr_area_kecil[i, j]
    offsets = [-0.2, 0, 0.2]
    colors = ['blue', 'green', 'red']
    for mn in range(3):
      nilai_satuan = nilai_bgr[mn]
      axes_2[1].text(j, i + offsets[mn], str(nilai_satuan),
                     ha='center', va='center', fontsize=8,
                     color=colors[mn], fontweight='bold')

axes_2[1].set_title('Zoom 8x8 piksel\nNilai [B, G, R]')
axes_2[1].axis('off')
plt.suptitle('Representasi RGB (3 Channels)', fontweight='bold')
plt.tight_layout()
plt.show()

print()
print("Shape:", img_bgr.shape)
print("Total Pixel:", total_pixel)
print("Ukuran bytes:", ukuran_bytes)
print("Ukuran MB:", f"{ukuran_mb:.2f} MB")
```
